```python
import pandas as pd
import requests
from urllib.parse import urlencode
import json, datetime
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
%matplotlib inline

mpl.style.use('seaborn')
api_key = "94bd224fdd23e740f91f7fc88375518f"

city = "Warsaw"
unit = "metric"
```

```python
params = dict(
    q=city,
    appid=api_key
)

url = 'http://api.openweathermap.org/data/2.5/forecast?q={}&units={}&APPID={}'.format(city, unit, api_key)
r = requests.get(url)
weather = json.loads(r.text)
df = pd.DataFrame.from_dict(weather['list'])
```

```python
from datetime import datetime

df["clouds"] = df.apply(lambda x: x.clouds["all"], axis=1)
df["rain"] = df.apply(lambda x: list(x.rain.values()), axis=1)
df["snow"] = df.apply(lambda x: list(x.snow.values()), axis=1)
df["sys"] = df.apply(lambda x: list(x.sys["pod"]), axis=1)
df["dt_txt1"] = df.apply(lambda x: datetime.strptime(x.dt_txt, '%Y-%m-%d %H:%M:%S'), axis=1)

df  = df.merge(df.wind.apply(lambda s: pd.Series({'wind_speed':s["speed"], 'wind_deg':s["deg"]})), 
    left_index=True, right_index=True)
df = df.merge(df.main.apply(lambda s: pd.Series({'temp':s["temp"], 'temp_min':s["temp_min"], 'temp_max':s["temp_max"], 'pressure': s['pressure'], "sea_level": s["sea_level"], "grnd_level": s["grnd_level"], "humidity": s["humidity"], "temp_kf": s["temp_kf"]})), 
    left_index=True, right_index=True)
df["wind_speed"] = df.apply(lambda x: x.wind_speed * 3.6, axis=1)
df = df.drop(columns=["main", "wind", "weather"])
date_df = df.loc[df["dt_txt1"].dt.day == 7]
```

```python
date_df
```

```python
def temp_odcz(temp, wind_speed):
    return 13.12 + 0.6215 * temp - 11.37 * wind_speed**0.16 + 0.3965 * temp * wind_speed**0.16

def temp_pkt_rosy(temp, humidity):
    return (humidity/100)**(1/float(8)) * abs(112 + (0.9 * temp)) + (0.1 * temp) - 112

pd.options.mode.chained_assignment = None  # default='warn'

date_df["temp_odcz"] = date_df.apply(lambda row: temp_odcz(row.temp, row.wind_speed), axis=1)
date_df["temp_odcz_min"] = date_df.apply(lambda row: temp_odcz(row.temp_min, row.wind_speed), axis=1)
date_df["temp_odcz_max"] = date_df.apply(lambda row: temp_odcz(row.temp_max, row.wind_speed), axis=1)

date_df["temp_pkt_rosy"] = date_df.apply(lambda row: temp_pkt_rosy(row.temp, row.humidity), axis=1)
date_df["temp_pkt_rosy_min"] = date_df.apply(lambda row: temp_pkt_rosy(row.temp_min, row.humidity), axis=1)
date_df["temp_pkt_rosy_max"] = date_df.apply(lambda row: temp_pkt_rosy(row.temp_max, row.humidity), axis=1)


date_df
```
